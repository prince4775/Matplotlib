```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties

plt.style.use("fivethirtyeight")

# Data and Labels
models = ['DNN[22]','CNN[6]', 'LSTM[15]', 'CNN-LSTM[21]', 'CNN-LSTM[5]',
          'CNN-BiLSTM[29]', 'Parallel CNN-LSTM[12]',
          'Multi Channel CNN-LSTM[23]', 'LSTM-SC[30]','TCN-BiLSTM[34]',
          'Proposed Model']

# Metrics Data
mape = [3.9, 3.94, 3.4, 3.6, 4.3, 3.9,  4, 0, 3.46, 3.95, 3.10]
mae = [574.57, 581.17, 521.3, 536.42, 636.9, 580.2, 590.2, 888.3, 518.81, 586.1, 302.27]
rmse = [772.3, 773.72, 742.8, 718.6, 824.1, 790.1, 817.7, 1140, 731.3, 776.02, 2191.78]

x = np.arange(len(models))  # X-axis positions
width = 0.2  # Narrower bar width for more spacing
spacing = 0.1  # Extra spacing between groups of bars

# Plotting Bars
fig, ax1 = plt.subplots(figsize=(12, 6))  # Set figure size

# Plot MAPE on Left Y-axis
bars1 = ax1.bar(x - width - spacing, mape, width, label='MAPE', color='lightskyblue', edgecolor='black')
ax1.set_ylabel('MAPE', fontsize=10, color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Add values for MAPE
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=7, color='black')

# Plot MAE on Right Y-axis
ax2 = ax1.twinx()  # Instantiate a second axes
bars2 = ax2.bar(x, mae, width, label='MAE', color='orange', edgecolor='black', hatch='//')
ax2.set_ylabel('MAE & RMSE', fontsize=10, color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add values for MAE
for bar in bars2:
    height = bar.get_height()
    ax2.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=7, color='black')

# Plot RMSE on Right Y-axis (same scale as MAE)
bars3 = ax2.bar(x + width + spacing, rmse, width, label='RMSE', color='#ec7063', edgecolor='black',hatch='\\')

# Add values for RMSE
for bar in bars3:
    height = bar.get_height()
    ax2.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=7, color='black')

# Customize Axes and Labels
font_properties = FontProperties(family='Arial', size=8)
ax1.set_xlabel('Models', fontsize=10)
ax1.set_title('24_hrs_Comparison of Proposed Model', fontsize=12)
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=45, fontproperties=font_properties)

# Add Boundary around the Graph
for spine in ax1.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)
for spine in ax2.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Add Legend Inside Graph
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9), ncol=3, fontsize=8)
plt.tight_layout()  # Adjust layout to avoid clipping

# Save the Graph
#plt.savefig('24_hrs_comparison_model_graph.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties

#plt.style.use("fivethirtyeight")

# Data and Labels
models = ['DNN[22]', 'CNN[6]', 'LSTM[15]', 'CNN-LSTM[21]', 'CNN-LSTM[5]',
          'CNN-BiLSTM[29]', 'Parallel CNN-LSTM[12]',
          'Multi Channel CNN-LSTM[23]', 'LSTM-SC[30]','TCN-BiLSTM[34]',
          'Proposed Model']

# Metrics Data
mape = [3.9, 3.94, 3.4, 3.6, 4.3, 3.9, 4, 6.24, 3.46, 3.95, 3.10]
mae = [574.57, 581.17, 521.3, 536.42, 636.9, 530.2, 590.2, 888.3, 518.81, 586.1, 302.27 ]
rmse = [772.3, 773.72, 742.8, 718.6, 824.1, 790.1, 817.7, 1140, 731.3, 776.02, 2196.78]

x = np.arange(len(models))  # X-axis positions
width = 0.2  # Narrower bar width for more spacing
spacing = 0.4  # Extra spacing between groups of bars

# Plotting Bars
fig, ax1 = plt.subplots(figsize=(12, 6))  # Set figure size

# Plot MAPE on Left Y-axis
bars1 = ax1.bar(x - width + spacing-0.45, mape, width, label='MAPE', color='lightskyblue', edgecolor='black')
ax1.set_ylabel('MAPE', fontsize=10, color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Add values for MAPE
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=7, color='black')

# Plot MAE on Right Y-axis
ax2 = ax1.twinx()  # Instantiate a second axes
bars2 = ax2.bar(x, mae, width, label='MAE', color='orange', edgecolor='black', hatch='//')
ax2.set_ylabel('MAE', fontsize=10, color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Add values for MAE
for bar in bars2:
    height = bar.get_height()
    ax2.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=7, color='black')

# Plot RMSE on Second Right Y-axis
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Shift third axis outward
bars3 = ax3.bar(x - width + spacing, rmse, width, label='RMSE', color='#ec7063', edgecolor='black', linestyle='--')
ax3.set_ylabel('RMSE', fontsize=10, color='black')
ax3.tick_params(axis='y', labelcolor='black')

# Add values for RMSE
for bar in bars3:
    height = bar.get_height()
    ax3.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=7, color='black')

# Customize Axes and Labels
font_properties = FontProperties(family='Arial', size=8)
ax1.set_xlabel('Models', fontsize=10)
ax1.set_title('Comparison of Proposed Model', fontsize=12)
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=45, fontproperties=font_properties)

# Add Boundary around the Graph
for spine in ax1.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)
for spine in ax2.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)
for spine in ax3.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Add Legend Inside Graph
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.8), ncol=3, fontsize=8)
plt.tight_layout()  # Adjust layout to avoid clipping
plt.grid()
# Save the Graph
#plt.savefig('comparison_model_graph.png', dpi=300, bbox_inches='tight')
plt.show()
```
