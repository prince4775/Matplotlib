## multistep

```python
data1_mape = [3, 2.7, 2.75, 4.36, 3.2, 4.15, 3.4, 3.3, 5.88, 2.7, 0, 2.84]
data2_mape = [3.9, 3.94, 3.4, 6.66, 3.6, 4.3, 3.9, 4, 6.24, 3.46, 3.95, 3.10]
```

```python
print(len(data1_mape))
print(len(data2_mape))
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties

plt.style.use("fivethirtyeight")


x = np.arange(len(models))  # X-axis positions
width = 0.4  # Bar width

# Plotting Bars
fig, ax1 = plt.subplots(figsize=(15, 8))  # Set figure size

# Plot MAPE for Dataset 1
bars1 = ax1.bar(x - width/2, data1_mape, width, label='ISO-NE_MAPE', color='#C9DFEC', edgecolor='black')

# Plot MAPE for Dataset 2
bars2 = ax1.bar(x + width/2, data2_mape, width, label='AEP_MAPE', color='#D5D6EA', edgecolor='black', hatch='//')

# Add values for Dataset 1 MAPE
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=12, color='black')

# Add values for Dataset 2 MAPE
for bar in bars2:
    height = bar.get_height()
    ax1.annotate(f'{height:.1f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # Offset text position
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=12, color='black')

# Customize Axes and Labels
font_properties = FontProperties(family='Arial', size=14)
ax1.set_xlabel('Models', fontsize=14)
ax1.set_title('Generalization_Multi_step', fontsize=15)
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=0, fontproperties=font_properties)
ax1.set_ylabel('MAPE', fontsize=12, color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Add Boundary around the Graph
for spine in ax1.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Add Legend Inside Graph
ax1.legend(loc='upper left', bbox_to_anchor=(0.01, 0.95), ncol=2, fontsize=10)
plt.tight_layout()  # Adjust layout to avoid clipping

# Save the Graph
plt.savefig(r'C:\Users\HP\OneDrive\Machine learning course\Bar_plots\paper1_genralization_Multi_step.png', dpi=300, bbox_inches='tight')
plt.show()
```
